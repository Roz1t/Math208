```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def plot_binomial_vs_normal_approximation(n, p):

    binomial_data = np.random.binomial(n, p, 1000)

    mean_binomial = np.mean(binomial_data)
    std_dev_binomial = np.std(binomial_data)

    mean_normal = n * p
    std_dev_normal = np.sqrt(n * p * (1 - p))

    plt.hist(binomial_data, bins='auto', density=True, alpha=0.5, color='blue', label='Binomial Distribution')

    x = np.linspace(mean_binomial - 4 * std_dev_binomial, mean_binomial + 4 * std_dev_binomial, 100)
    y = norm.pdf(x, mean_normal, std_dev_normal)
    plt.plot(x, y, 'r', label='Normal Distribution Approximation')

    plt.title('Binomial vs. Normal Distribution Approximation')
    plt.xlabel('Value')
    plt.ylabel('Probability Density')
    plt.legend()

    plt.show()

n = 100
p = 0.2

if n * p > 5 and n * (1 - p) > 5:
    plot_binomial_vs_normal_approximation(n, p)
    print("Conditions np > 5 and nq > 5 are satisfied.")
    print("When n = 100 and p = 0.2")
else:
    print("Conditions np > 5 and nq > 5 are not satisfied.")

```
