```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import t

np.random.seed(42)
random_numbers = t.rvs(df=10, size=100)
original_mean = np.mean(random_numbers)
original_std = np.std(random_numbers)

sample_means = []
sample_stds = []

for _ in range(15):
    sample_group = np.random.choice(random_numbers, size=30, replace=False)
    sample_mean = np.mean(sample_group)
    sample_std = np.std(sample_group)

    sample_means.append(sample_mean)
    sample_stds.append(sample_std)

print(f"Original mean (μ): {original_mean:.4f}, Original std (σ): {original_std:.4f}")
print(f"Sample mean (μ): {np.mean(sample_means):.4f}, Sample std (σ): {np.mean(sample_stds):.4f}")

plt.hist(sample_means, bins=15, edgecolor='black', density=True)
plt.title('Distribution of Sample Means')
plt.xlabel('Sample Mean')
plt.ylabel('Frequency')
plt.show()
```
